# Starter Exploratory Data Analysis Notebook 

Here is a starter notebook to look at exploratory data analysis and preprocessing. More can be added and explored ! 

```python
import pandas as pd
```

```python
col_names = ['Time','Tweet','Favorites','Retweets','User','UserName','Follower Count','Friend Count', 'User Given Location', 'Place','Coordinates']
```

```python
df = pd.read_csv('HashtagGreenNewDeal.csv', names = col_names, index_col = 'User' )
```

```python
df.head()
```

```python
df.Tweet.head()
```

```python
df = df.astype('object')

for index,row in df.iterrows():
    mentions = []
    urls = []
    hashtags = []
    split_tweet_array = row['Tweet'].split()
    for word in split_tweet_array:
        if word.startswith('b\'@'):
            mentions.append(word[3:])
        if word.startswith('@'):
            mentions.append(word[1:])
        if word.startswith('https:'):
            urls.append(word) 
        if word.startswith('#'):
            hashtags.append(word[1:])
    df.at[index,'Mentions'] = mentions
    df.at[index,'URL'] = urls
    df.at[index,'Hashtags'] = hashtags
  
    
```

```python
df.head()
```

```python
# Apply a first round of text cleaning techniques
import re
import string

def clean_text_round1(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    text = text[2:]
    #text = re.sub('\[.*?\]', '', text) #remove all information within square brackets (applause queues)
    #text = re.sub('b\'' , '') #remove all bs at the start of tweets
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) #get rid of all punctuation
    text = re.sub('\w*\d\w*', '', text) #\d is all digits, w is alphanumeric characters * is 0 or more times 
    return text

round1 = lambda x: clean_text_round1(x)
```

```python
data_clean = pd.DataFrame(df.Tweet.apply(round1))
data_clean
```

```python
# Apply a second round of cleaning
def clean_text_round2(text):
    '''Get rid of some additional punctuation and non-sensical text that was missed the first time around.'''
    text = re.sub('[‘’“”…]', '', text) #get rid of all the quotes 
    text = re.sub('\n', '', text) #get rid of line breaks \n
    return text

round2 = lambda x: clean_text_round2(x)
```

```python
data_clean = pd.DataFrame(data_clean.Tweet.apply(round2))
data_clean
```

```python
df['Tweet'] = data_clean['Tweet']
```
